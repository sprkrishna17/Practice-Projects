**Oversampling method for balancing of the Dataset**

```python
#Checking value count of target column
y.value_counts()
```

```python
#oversampling the data
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()
```

Here we can see that both 0 & 1 has equal number of values we use this to prevent the biaseness in the dataset

**Checking Variance Inflation factor(VIF)**

```python
#Finding varience inflation factor in each scaled column

from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['VIF Values']=[variance_inflation_factor(x.values,i)
            for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

As the VIF value of features are less than 5, we can say that there is no multicolinearity present in the dataset so no need to remove the multicolinearity

# Modeling

Since the outcome has only two variable we will use binary clasification model

**Finding best random state**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score
```

```python
for i in range (0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1)==round(accuracy_score(y_test,pred_test)*100,1):
        print(f"At random state {i}, The model performs very well")
        
        print(f"At random state {i}, the training accuracy is:- {accuracy_score(y_train,pred_train)*100}")
        print(f"At random state {i}, the testing accuracy is:- {accuracy_score(y_test,pred_test)*100}")
        print("\n")
```

Best training accuracy is 79.5%, best testing accuracy is 79.5% on Random_state 447

# Creating Train-Test-split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .20, random_state = 447)
```

```python
print('Training Survived - \n', y_train.value_counts())
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train, y_train)
predlr = LR.predict(x_test)
print("Accuracy",accuracy_score(y_test, predlr)*100)
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

**for Logistic Regression, from above Confusion matrix, TP = 70, FP = 16, FN = 25, TN = 89. accuracy = 79.5%.**

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
preddt = dt.predict(x_test)
print("Accuracy",accuracy_score(y_test, preddt)*100)
print(confusion_matrix(y_test,preddt))
print(classification_report(y_test,preddt))
```


**for Decision Tree Classifier, from above Confusion matrix, TP= 60, FP = 26, FN = 34, TN = 80. accuracy = 70%.**

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
predrf = dt.predict(x_test)
print("Accuracy",accuracy_score(y_test, predrf)*100)
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

**for Random Forest Classifier, from above Confusion matrix, TP = 60, FP = 26, FN = 34, TN = 80. accuracy = 70%.**

# SVM

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit( x_train, y_train)

ad_pred = svc.predict(x_test)
print("Accuracy", accuracy_score(y_test, ad_pred)*100)
print(confusion_matrix(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
```

**for SVM, from above Confusion matrix, TP = 73, FP = 13, FN = 28, TN =86. accuracy = 79.5%.**

```python
GCV_pred=GCV.best_estimator_.predict(x_test) # Prediciting with the best parameters
accuracy_score(y_test,GCV_pred) # Checking final accuracy
```
